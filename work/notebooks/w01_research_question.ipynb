## 2. The question: decision, action, cost of a wrong call

**Research question (one sentence):** Which content pages should be reviewed first for refresh, expansion, protection, or monitoring, given limited reviewer capacity?

| Element | My answer |
|---|---|
| **Decision improved** | Prioritizing the daily/weekly content review queue — who gets scarce editor and SEO analyst time |
| **Unit of analysis** | One row = one content page (90-day observable window, pseudonymized `content_id`) |
| **Output** | Ranked queue with scores, suggested action, and reason codes |
| **Action someone takes** | A content or SEO reviewer opens the top-K list, inspects reason codes, and chooses refresh / rewrite / monitor / protect |
| **Cost of a wrong recommendation** | **False positive:** editor spends hours refreshing a page that was noise or seasonal — wasted capacity. **False negative:** a high-impression declining page is missed and keeps losing visibility — opportunity cost. This is decision-support, not autonomous publishing. |

**Why data/ML helps:** The inventory is too large to review manually page-by-page. Observable signals (impressions, position, trend, age, engagement) vary across pages; a ranked model can surface the highest-evidence candidates faster than a fixed rule — if validated honestly.

## 3. Quick look at the data (2-3 real numbers)

Three numbers from the starter dataset and pipeline that make Lane 2 worth the next 7 weeks — computed live below from `data/raw/content_refresh_anonymized.csv` and `outputs/model_results.json`.

```python
import json
import os
from pathlib import Path

import pandas as pd

# Find repo root (works locally and in Colab after clone)
root = Path.cwd()
while root != root.parent and not (root / "data" / "raw").exists():
    root = root.parent
os.chdir(root)

raw = pd.read_csv("data/raw/content_refresh_anonymized.csv")
results = json.load(open("outputs/model_results.json"))

n_rows = len(raw)
declining_pct = (raw["trend_direction"] == "down").mean() * 100
high_vis = (raw["impressions_90d"] >= 500).sum()
baseline_p50 = results["baseline"]["baseline_precision_at_50"]
rf_p50 = results["models"]["random_forest"]["precision_at_50"]

print(f"1) Starter slice size: {n_rows:,} content pages (after repo filters: impressions>0, age>=90d)")
print(f"2) Share trending down: {declining_pct:.1f}% — large enough class to rank, not a rare edge case")
print(f"3) High-visibility pages (impressions_90d >= 500): {high_vis:,} — these are the pages worth reviewer time")
print(f"4) Precision@50 — baseline rule: {baseline_p50:.2f} (~{round(baseline_p50*50)} of top 50 correct)")
print(f"5) Precision@50 — random forest: {rf_p50:.2f} (~{round(rf_p50*50)} of top 50 correct)")
print(f"   → Learned ranking beats the hand-written rule by ~{rf_p50/baseline_p50:.1f}x on this metric")
print(f"   Validation: {results['split_strategy']} (clients held out of training)")
```
